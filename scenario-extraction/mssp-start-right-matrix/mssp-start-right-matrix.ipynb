### lc 범위 내에서, rc 개수만큼의 노드를 중복 없이 생성

```python
import numpy as np
import random

lc = 1134890
rc = 100

lc_list = np.arange(lc).tolist()
# lc 범위 내에서, rc 개수만큼의 노드를 중복 없이 생성
lc_index = random.sample(lc_list,  rc)
```

## MSSP 시 사용될 우측 행렬 생성

### col index 는 오름차순
### row index 는 중복 없이 생성한 노드 번호
### element 는 1

```python
outputfile = "/Users/bdlab/Desktop/com-youtube-right-matrix-100.txt"

# Output : Right Matrix
with open(outputfile,"a+") as wf:

    # col 을 오름차순 탐색
    for col in range(rc):

        # COO 형식의 데이터 생성
        # col 은 rc index
        # row 는 lc index
        # element 는 1
        wf.write(str(lc_index[col]) + " " + str(col) + " " + str(1) + "\n")
```
